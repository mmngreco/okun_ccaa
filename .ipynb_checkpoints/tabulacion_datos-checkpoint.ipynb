```python
df_empleo = DataFrame(vstack((empleo[1].ix[:101,:], empleo[0])), columns=empleo[0].columns)
```

```python
def trimestre(x):
    return x[x.index('T'):]
```

```python
def romans_to_num(x):
    roman = x[x.index('T') + 1:]
    translate = {'I': 1, 'II': 2, 'III': 3, 'IV': 4}

    if any(map(lambda y: roman == y, translate.keys())):
        return x[:x.index('T')+1] + str(translate[roman])
    else:
        return x
```

```python
def year(x):
    return x[x.index('T')-4:x.index('T')]
```

```python
df_empleo['trimestre'] = df_empleo.Tiempo.apply(trimestre).apply(romans_to_num)
```

```python
df_empleo['Tiempo'] = df_empleo.Tiempo.apply(romans_to_num)
```

```python
df_empleo.head()
```

```python
del df_empleo['trimestre']
```

```python
t = df_empleo.Tiempo.copy(deep=True)
```

```python
del df_empleo['Tiempo']
```

```python
tasa_empleo = df_empleo.pct_change(periods=4)
```

```python
tasa_empleo['tiempo'] = t
```

```python
tasa_empleo[u'año'] = t.apply(year)
```

```python
tasa_empleo.groupby(u'año')
```

```python
tasa_empleo.groupby(u'año').mean().dropna(axis=0).to_excel('d_empleo.xls')
```
