```python
def trans_decimal(x):
    try:
        return float(x.replace('.', '').replace(',', '.')) 
    except:
        return x
```

```python
def trimestre(x):
    return x[x.index('T'):]
```

```python
def romans_to_num(x):
    roman = x[x.index('T') + 1:]
    translate = {'I': 1, 'II': 2, 'III': 3, 'IV': 4}

    if any(map(lambda y: roman == y, translate.keys())):
        return x[:x.index('T')+1] + str(translate[roman])
    else:
        return x
```

```python
def year(x):
    return x[x.index('T')-4:x.index('T')]
```
