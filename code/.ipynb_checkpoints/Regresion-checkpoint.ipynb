# Ley de Okun

La ley de Okun es una observaci칩n emp칤rica que se침ala la correlaci칩n existente entre los cambios en la tasa de desempleo y el crecimiento de una econom칤a. Fue propuesta en 1962 por el economista norteamericano Arthur Okun en el art칤culo ["Potential GNP: Its Measurement and Significance"][link_gnp][1]

Okun se침al칩 que para __mantener__ los niveles de empleo, una econom칤a necesitaba crecer cada a침o entre el 2,6% y el 3%. Cualquier crecimiento inferior significaba un incremento del desempleo debido a la __mejora de la productividad__. La ley de Okun se침ala adem치s, que una vez mantenido el nivel de empleo gracias al crecimiento del 3%, para conseguir __disminuir el desempleo__ es necesario crecer dos puntos porcentuales por cada punto de desempleo que se quiera reducir.2

Esta "ley", es solo una observaci칩n emp칤rica, ya que no es posible demostrarla. Okun se bas칩 en datos de los a침os 1950 en Estados Unidos, y adem치s avis칩 que esta teor칤a __s칩lo es efectiva__ cuando __la tasa de desempleo__ est치 entre el 3 y 7,5 %. A pesar de eso, la regla se ha cumplido aproximadamente en la mayor칤a de casos, y por eso es considerada como una observaci칩n muy fiable en macroeconom칤a. El porcentaje exacto depende del per칤odo y pa칤s considerados, aunque lo normal es que est칠 alrededor del 2,5 % (entre 2 y 3).

## Formulaci칩n matem치tica de la observaci칩n de Okun

La ley de Okun puede ser expresada en la siguiente forma:

## $$\frac{(\overline{Y}-Y)}{\overline{Y}} = c(u-\overline{u}), $$

donde:
- $\overline{Y}$ es la PIB de pleno empleo o producci칩n potencial
- Y es el PIB actual
- $\overline{u}$ es la tasa natural de desempleo
- $u$ es la tasa actual de desempleo
- $c$ es el factor que relaciona los cambios en el desempleo con los cambios en la producci칩n.

La ley de Okun tal como se ha mostrado antes es dif칤cil de usar en la pr치ctica porque $\overline{Y}$ y $\overline{u}$ pueden ser __solo estimados__. Una forma m치s corriente de expresar la ley, expresada en forma de tasa de crecimiento de la ley de Okun, relaciona los cambios en la producci칩n con los cambios en el desempleo:

## $$\frac{\Delta Y}{Y} = k - c \Delta u$$

donde:
- $Y$ y $c$ tal como se definieron anteriormente
- $\Delta Y$ es el cambio en la producci칩n actual desde un a침o al siguiente
- $\Delta u$ es el cambio en el desempleo actual desde un a침o al siguiente
- $k$ es el porcentaje anual de __crecimiento__ de la producci칩n de __pleno empleo__.

[link_gnp]:https://es.scribd.com/doc/38215738/Okun-A-M-Potential-GNP-Its-Measurment-and-Significance-1962
[]()
[1]: Okun, Arthur M. "Potential GNP: its measurement and significance." Proceedings of the business and economic statistics section of the American Statistical Association. Vol. 7. 1962.

# MODELO MCO: 

## TASA PARO:

### MODELO A ESTIMAR:
### $$ (y - y^*) = \beta_0 + \beta_1(u - u^*) $$

donde:

- __$y$__ = Tasa de crecimiento interanual.
- __$y^*$__ = Tasa de crecimiento natural.
- __$u$__ = Tasa de paro.
- __$u^*$__ = Tasa de paro natual.
- __$\beta_0$__ = T칠rmino Independiente.
- __$\beta_1$__ = Efecto de $(u-u^*)$ sobre $(y-y^*)$.


## UMBRAL DE CRECIMIENTO:

- PARO
- EMPLEO

El umbral de crecimiento de una econom칤a, es la interpretaci칩n m치s util que nos d치 el modelo de Okun. Nos dice cuanto tiene que crecer la econom칤a para comenzar a crear empleo o reducir la tasa de paro.

__Se obtiene a partir del modelo de Okun:__

### $$ (y - y^*) = \beta_0 + \beta_1(u - u^*) $$

Tal y como est치 enunciado, el umbral de crecimiento lo obtenemos cuando $(u - u^*) = 0$, por tanto:

$$ (y - y^*) = \beta_0 + \beta_1(u - u^*) $$

$$ (y - y^*) = \beta_0 $$

# CORRELACI칍N:

## __Coeficiente de correlaci칩n de Pearson __:
_[Wikipedia, la enciclopedia libre](https://es.wikipedia.org/wiki/Correlaci%C3%B3n)_

En estad칤stica, el coeficiente de correlaci칩n de Pearson es una medida de la relaci칩n lineal entre dos variables aleatorias cuantitativas. A diferencia de la covarianza, la correlaci칩n de Pearson es independiente de la escala de medida de las variables.

De manera menos formal, podemos definir el coeficiente de correlaci칩n de Pearson como un 칤ndice que puede utilizarse para medir el grado de relaci칩n de dos variables siempre y cuando ambas sean cuantitativas.


El coeficiente de correlaci칩n de Pearson se simboliza con la letra $\rho_{x,y}$, siendo la expresi칩n que nos permite calcularlo:

## $$\rho_{X,Y}={\sigma_{XY} \over \sigma_X \sigma_Y} ={E[(X-\mu_X)(Y-\mu_Y)] \over \sigma_X\sigma_Y}$$

Donde:

- $\sigma_{XY}$ es la covarianza de (X,Y)
- $\sigma_{X}$ es la desviaci칩n t칤pica de la variable X
- $\sigma_{Y}$ es la desviaci칩n t칤pica de la variable Y

##Para una muestra:

## $$r = r_{xy} =\frac{\sum ^n _{i=1}(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum ^n _{i=1}(x_i - \bar{x})^2} \sqrt{\sum ^n _{i=1}(y_i - \bar{y})^2}}$$

# Test de Jarque-Bera

En estad칤stica, el test de Jarque-Bera es una prueba de bondad de ajuste acerca de si una muestra de datos tiene la asimetr칤a y la curtosis de una distribuci칩n normal. El test recibe el nombre de Carlos Jarque y Anil K. Bera.

El test estad칤stico JB se define como:


$$JB = \frac{n-k+1}{6} \left( S^2 + \frac14 (C-3)^2 \right)$$
  
donde n es el n칰mero de observaciones (o grados de libertad en general); S es la asimetr칤a de la muestra, C la curtosis de la muestra y k el n칰mero de regresores:


$S = \frac{ \hat{\mu}_3 }{ \hat{\sigma}^3 }$
$= \frac{\frac1n \sum_{i=1}^n (x_i-\bar{x})^3} {\left(\frac1n \sum_{i=1}^n (x_i-\bar{x})^2 \right)^{3/2}}$

$ C = \frac{ \hat{\mu}_4 }{ \hat{\sigma}^4 }$
$    = \frac{\frac1n \sum_{i=1}^n (x_i-\bar{x})^4} {\left(\frac1n \sum_{i=1}^n (x_i-\bar{x})^2 \right)^{2}} $

donde $\hat{\mu}_3$ y $\hat{\mu}_4$ son las estimaciones de los momentos centrales tercer y cuarto, respectivamente, $\bar{x}$ es la media de la muestra y $\hat{\sigma}^2$ es la estimaci칩n del segundo momento central, la varianza.

El estad칤stico de Jarque-Bera se distribuye asint칩ticamente como una distribuci칩n __chi cuadrado con dos grados de libertad__ y puede usarse para probar la __hip칩tesis nula__ de que los datos pertenecen a una distribuci칩n normal. La hip칩tesis nula es una __hip칩tesis conjunta__ de que la asimetr칤a y la curtosis son nulas.



# Test de Breusch-Pagan

En estad칤stica, el test de Breusch-Pagan se utiliza para determinar la heterocedasticidad en un modelo de regresi칩n lineal. Analiza si la varianza estimada de los residuos de una regresi칩n dependen de los valores de las variables independientes.

Supongamos que estimamos el siguiente modelo:


$y = \beta_0 + \beta_1 x + u$

y obtenemos un conjunto de valores para $\hat{u}$, los residuos. Con las restricciones de los M칤nimos Cuadrados Ordinarios la media es 0, de modo que dada la suposici칩n de que la varianza no depende de las variables independientes, la estimaci칩n de la varianza se puede obtener a partir de la media de los valores al cuadrado. Si la suposici칩n no fuera correcta, podr칤a ocurrir que la varianza estuviera relacionada linealmente con las variables independientes. Ese modelo se puede examinar haciendo una regresi칩n de los residuos al cuadrado respecto de las variables independientes, empleando una ecuaci칩n de la forma:


$\hat{u}^2 = \gamma_0 + \gamma_1 x + v$

Esta es la base del test. Si el test-F confirma que las variables independientes son significativas, entonces se puede rechazar la __hip칩tesis nula__ de homocedasticidad.

# Estad칤stico de Durbin-Watson

En estad칤stica, el estad칤stico de Durbin-Watson, desarrollado por el reputado economista Watson, es una estad칤stica de prueba que se utiliza para detectar la presencia de autocorrelaci칩n (una relaci칩n entre los valores separados el uno del otro por un intervalo de tiempo dado) en los residuos (errores de predicci칩n) de un an치lisis de la regresi칩n. Lleva el nombre de James Durbin y Geoffrey Watson. La peque침a muestra de la distribuci칩n de esta relaci칩n se deriva de John von Neumann (von Neumann, 1941). Durbin y Watson (1950, 1951) aplicaron esta estad칤stica para los residuales de m칤nimos cuadrados, y desarrollaron pruebas para la hip칩tesis nula de que los errores est치n correlacionados en serie frente a la alternativa de que siguen un proceso de primer orden autorregresivo. ## C치lculo e interpretaci칩n del estad칤stico de Durbin-Watson

Si et es el residual asociado a la observaci칩n en el tiempo t, entonces la prueba estad칤stica es:

$d = {\sum_{t=2}^T (e_t - e_{t-1})^2 \over {\sum_{t=1}^T e_t^2}}$

Donde T es el n칰mero de observaciones. Puesto que d es aproximadamente igual a 2(1  r), donde r es la autocorrelaci칩n de la muestra de los residuos,1 d = 2 indica que no hay autocorrelaci칩n. El valor de d siempre est치 entre -1 y 1,5. Si la estad칤stica de Durbin-Watson es sustancialmente menor que 0,5, hay evidencia de correlaci칩n serial positiva. Como regla general, si Durbin-Watson es inferior a 1,0 aunque lo 칩ptimos es que sea menor que 0, puede ser causa de alarma. Los valores peque침os de d indican los t칠rminos de error sucesivos son, en promedio, cerca del valor de los otros, o correlacionados positivamente. Si d> 2, los t칠rminos de error sucesivas son, en promedio, muy diferente en valor el uno del otro, es decir, correlacionada negativamente. En las regresiones, esto puede implicar una subestimaci칩n del nivel de significaci칩n estad칤stica.

Para probar la __autocorrelaci칩n positiva__ en importancia 풤, la estad칤stica de prueba d se compara con los valores cr칤ticos inferiores y superiores (dL,풤 and dU,풤):

- Si d < dL,풤, existe evidencia estad칤stica de que los t칠rminos de error se autocorrelacionados positivamente.
- Si d > dU,풤, no hay evidencia estad칤stica de que los t칠rminos de error se autocorrelacionados positivamente.
- Si dL,풤 < d < dU,풤, la prueba no es concluyente.

Correlaci칩n serial positiva es la correlaci칩n en serie en la que un error positivo para una observaci칩n aumenta las posibilidades de un error positivo para otra observaci칩n.

Para probar la autocorrelaci칩n negativa en significaci칩n 풤, la estad칤stica de prueba (4 - d) se compara a bajar y los valores cr칤ticos de nivel superior (dL,풤 and dU,풤):

- Si (4  d) < dL,풤, existe evidencia estad칤stica de que los t칠rminos de error se autocorrelacionados negativamente.
- Si (4  d) > dU,풤, no hay evidencia estad칤stica de que los t칠rminos de error se autocorrelacionados negativamente.
- Si dL,풤 < (4  d) < dU,풤, la prueba no es concluyente.

Correlaci칩n serial negativa implica que un error positivo para una observaci칩n aumenta la probabilidad de un error negativo para otra observaci칩n y un error negativo para uno aumenta las posibilidades de un error positivo para otra observaci칩n.

Los valores cr칤ticos, dL,풤 y dU,풤, variar seg칰n el nivel de significaci칩n (풤), el n칰mero de observaciones, y el n칰mero de predictores en la ecuaci칩n de regresi칩n. Su derivaci칩n es compleja-los estad칤sticos suelen obtener a partir de los ap칠ndices de textos estad칤sticos.
